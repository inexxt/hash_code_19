```python
l = []
fnames = ! ls ../data
```

```python
fnames = fnames[1:]
```

```python
for fname in fnames:    
    with open("../data/" + fname) as f:
        l.append(f.readlines())
```

```python
[len(x) for x in l]
```

```python
from tqdm import tqdm
from collections import Counter

tt = []
for x in l[1:]:
    tags = Counter()
    for t in tqdm(x[1:]):
        tags = tags + Counter(list(t.split())[2:])
    print(len(tags))
    tt.append(tags)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

tags = tt[2]
xs, ys = zip(*sorted(tags.items(), key=lambda x: x[1]))
```

```python
plt.figure(figsize=(20, 20))
plt.barh(xs, ys)
plt.show()
```

```python
from tqdm import tqdm
import numpy as np

for x in l:
    num_tags = []
    for t in x[1:]:
        num_tags.append(len(list(t.split()[2:])))
    print(f"Avg {(sum(num_tags) / len(num_tags))} \t Med {np.median(num_tags)} \t Std {np.std(num_tags)} Max {(np.max(num_tags))}")
```
